```python
#advanced visualization revised 
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sb
import os
```

```python
os.getcwd()
```

```python
movies = pd.read_csv("P4-Movie-Ratings.csv")
```

```python
movies
```

```python
movies.columns
```

```python
movies.columns = ['Film','Genre', 'Criticsrating', 'Audienceratings', 'Budgets', 'Release']
```

```python
movies[2:5]
```

```python
movies
```

```python
movies.Genre = movies.Genre.astype('category')
movies.Film = movies.Film.astype('category')
movies.Release = movies.Release.astype('category')
```

```python
movies[2:5]
```

```python
movies.info()
```

```python
for gen in movies.Genre.cat.categories :
    print(gen)
```

```python
list1 = list()
mylabels = list()
for gen in movies.Genre.cat.categories :
    list1.append(movies[movies.Genre== gen].Budgets)
    mylabels.append(gen)

h = plt.hist(list1,bins=30,stacked=True , rwidth=1 , label=mylabels)
plt.legend()
plt.show()
```
