```python
x = np.array([1,2,3,4,5,6,7])
y = []

for i in range(len(x)):
    y.append(i*i)
    
colors  = ['red']*(len(x)-1)
colors.append('yellow')

plt.figure()
plt.scatter(x,y,s=100, c=colors)
```

```python
x1 = [2001,2006,2009,2014,2017]
y1 = [1,4,9,16,25]


plt.figure()
plt.scatter(x1[:1],y1[:1],s=50,c='red', label="NRI")
plt.scatter(x1[1:],y1[1:],s=50,c='yellow', label="Indians")

plt.xlabel('Years')
plt.ylabel('No. of People')
plt.legend()
```

```python
x2 = np.array([0,10,20,30,40,50])
y2 = 5*x2

plt.figure()
plt.plot(x2, '-o', y2, '-o')
plt.title("MI Vs SRH")
plt.xlabel('Overs')
plt.ylabel('Runs')
plt.legend(['MI','SRH'])
```

```python
x2 = np.array([0,10,20,30,40])
y2 = 5*x2

plt.figure()
plt.plot(x2, '-o', y2, '-o')
plt.gca().fill_between(range(len(x2)),x2,y2,facecolor='yellow',alpha=0.4)
```

```python
data = np.array([3,4,1,5,6,7])
x = range(len(data))

plt.bar(x,data,width=0.2,color='lime')
```

```python
datas = np.array([3,4,1,5,6,7])
newdata = []
for i in datas:
    newdata.append(i-0.5)
    
plt.figure()
plt.bar(datas,newdata,width=0.3,color='green')
```

```python
xrange = range(len(datas))
plt.figure()
plt.barh(xrange, datas, height=0.3, color='orange')
plt.barh(xrange, newdata, height=0.3, left=data, color='grey')
```

```python

plt.figure()
xvalue = ([1,2,3,4,5,6])
plt.bar(xvalue,datas,width=0.3,color='lime',label='Unacademy')

new_xvalue = []
for i in xvalue:
    new_xvalue.append(i + 0.3)
    
plt.bar(new_xvalue,newdata,width=0.3,color='purple',label='BYJU')
plt.legend()
```
